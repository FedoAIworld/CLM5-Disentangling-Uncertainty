```python
def convert_to_per_day(df, site, var, sim_var):
    rename_mapping = {f'{sim_var}': f'{site}_{var}'}
    #df[[f'{sim_var}']] = df[[f'{sim_var}']].mul(86400)
    df = df.rename(columns=rename_mapping)
    return df

def process_site_realizations(site_dfs, sites, var, sim_var, output_dir):
    site_realizations = {}
    
    # Ensure the output directory exists
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    
    for site, site_df in zip(sites, site_dfs):
        ens_df = {}
        for realization, df in site_df.items():
            ens_df[realization] = convert_to_per_day(df, site, var, sim_var)
        realizations = {}
        for data, df in ens_df.items():
            resampled_df = timeseries_custom_resample(df, 'datetime', 'MS')
            realizations[data] = resampled_df
            
            ## Save each realization DataFrame as a CSV file
            #csv_filename = f"{site}_realization_{data}.csv"
            #csv_filepath = os.path.join(output_dir, csv_filename)
            #resampled_df.to_csv(csv_filepath, index=True)
        
        site_realizations[site] = realizations
        
        # Create an aggregated DataFrame for the site with all realizations
        aggregated_df = pd.concat(
            [df.set_index('datetime') for df in realizations.values()], axis=1
        )
        aggregated_df.columns = [f"real_{i}" for i in range(aggregated_df.shape[1])]
        
        # Reset index to include 'datetime' as the first column
        aggregated_df.reset_index(inplace=True)
        
        # Save the aggregated DataFrame as a CSV file
        aggregated_csv_filename = f"{site}_aggregated.csv"
        aggregated_csv_filepath = os.path.join(output_dir, aggregated_csv_filename)
        aggregated_df.to_csv(aggregated_csv_filepath, index=False)
    
    return site_realizations
```
